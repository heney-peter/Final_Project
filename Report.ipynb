# Ski Resort Analysis Report
### Peter Heney

    The purpose of this project is to provide information useful to a skier that is deciding which pass they may be interested in purchasing for the upcoming season. There are two major international ski passes: Epic and Ikon. This involved creating an interactive map and graphs that users can access with no programming knowledge, as well as numerous data manipulations required to make the visuals accurate and consistent.

    The data I worked with came in two separate sets from the same source. First, 'skiresorts.csv' contains a list of ski resorts that are included in either the Epic or Ikon pass as well as summary information about mountain size, standard snow conditions, and the pass that each mountain belongs to. This is the data set that was used for mapping resorts. Resorts are located in the US, Canada, and Europe, so I converted all height measures (such as vertical drop and average snowfall) to US customary units for an easier comparison. Additionally, I used unique resort names to pull geographical location data for each resort so that they could be mapped. The second dataset, 'annualsnow.csv', has the same resorts listed with rows for each season (2012-Present) and the annual snowfall for each resort/year combination. I again converted metric measures of snowfall to US customary for consistency. 

    The end product for this data is an interactive dashboard with two applications. First, the user encounters a world map with each resort marked at its location. The user can select which pass they are interested in looking at and scroll around the world to see which resorts they can access with that pass. Hovering over any resort provides the user with its name, vertical drop, average annual snowfall, and the amount of skiable terrain. A user can look at each pass and zoom in on their area to see if that pass would be a good fit for them based on local mountain availability and size. The second feature is a graph of the past decade of annual snowfalls for a selected resort. The user can choose from a list of included resorts and see how the winters have been trending. For powder-chasers, just choose a mountain and see if they have been trending towards deep winters or a short season.

    Findings are highly dependent on the location of the user. For example, a quick scroll through the map shows that European users that intend to ski in the Alps should purchase an Epic pass. Mountain availability for Epic in that region is high with tons of options, while the Ikon view is barren. However, a skier in northern Michigan would see that Ikon provides them with two local options while Epic has none. Similarly, looking at the historical snowfall graphs provides very different outlooks depending on the resort. For a skier at Big Sky, Montana, it is probably worrisome that the past 3 years have been the lowest of the decade (all under 200 inches). Crystal Mountain, Washington, however, has seen a pretty steady 300-400 inches for the past 7 years. This would be a more reliable mountain for somebody planning an expensive trip and wanting a good chance of snow coverage.
